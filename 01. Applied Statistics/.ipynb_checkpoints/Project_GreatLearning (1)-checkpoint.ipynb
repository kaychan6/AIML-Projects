4.Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

```python
def Goal_diff_count(GoalsFor,GoalsAgainst):
    return GoalsFor - GoalsAgainst
```

```python
Laliga
```

```python
df1 = Laliga[['Team','GoalsFor','GoalsAgainst']]
df1
```

```python
Laliga['GoalsFor'] = Laliga.GoalsFor.astype(int)
Laliga['GoalsAgainst'] = Laliga.GoalsAgainst.astype(int)
```

```python
Laliga_df1 = Laliga[['Team','GoalsFor','GoalsAgainst']]
#Laliga_df1
```

```python
#Applying the Goal_diff_count function to each element in the dataset
Laliga_df1['Goal_diff'] = Laliga_df1.apply(lambda x: Goal_diff_count(x['GoalsFor'], x['GoalsAgainst']),axis=1)
```

```python
Laliga_df1
```

```python
Laliga_df1['Goal_diff'].min()
```

```python
Laliga_df1['Goal_diff'].max()
```

The team which has maximum and minimum goal difference.

```python
#Laliga_df1['Goal_diff']==Laliga_df1['Goal_diff'].max() #Returns True value for maximum value
Laliga_df1[Laliga_df1['Goal_diff']==Laliga_df1['Goal_diff'].max()]
```

```python
Laliga_df1[Laliga_df1['Goal_diff']==Laliga_df1['Goal_diff'].min()]
```

# Task 5


Create a new column with name “Winning Percent” and append it to the data set 
#Percentage of Winning = (GamesWon / GamesPlayed)*100
#If there are any numerical error, replace it with 0%
#Print the top 5 teams which has the highest Winning percentage

```python
Laliga['GamesWon'] = Laliga.GamesWon.astype(int)
Laliga['GamesPlayed'] = Laliga.GamesPlayed.astype(int)
```

```python
Winning_percent = (Laliga['GamesWon']/Laliga['GamesPlayed'])*100
Winning_percent[56:61]
```

Creating new column called Winning_percent and append it to the data set Laliga

```python
Laliga['Winning_Percent'] = (Laliga.loc[:,'GamesWon']/Laliga.loc[:,'GamesPlayed'])*100
Laliga.tail()
```

```python
Laliga['Winning_Percent']=Laliga['Winning_Percent'].fillna(value=0)
```

Numerical error 'NaN' is replacing with 0%

```python
Laliga.tail()
```

```python
Laliga.sort_values(by='Winning_Percent',ascending=False)
```

```python
(Laliga.sort_values(by='Winning_Percent',ascending=False)).head()
```

```python
df_task5 = (Laliga.sort_values(by='Winning_Percent',ascending=False)).head()
```
